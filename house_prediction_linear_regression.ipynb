```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

# loading data_set with pandas
df = pd.read_csv(r"C:\Users\indur\Downloads\house_prices.csv")

#cleaing data set like missing_value-------------------------------------------

df = df.dropna(subset=['Bedrooms','Bathrooms','Size_sqft','SalePrice'])
df

#converting string to numeric by usinf labelencoder cloumn is country--------------------------------

label = LabelEncoder()
df['Country'] = label.fit_transform(df['Country'])

feature = df[['Country','Bedrooms','Bathrooms','Size_sqft']]
target = df['SalePrice']

#train_test_split-----------------------------------------------------------------------

x_train,x_test,y_train,y_test = train_test_split(feature,target,test_size = 0.2,random_state = 42)

# model ----------------------------------------------------------------------------------

model = LinearRegression()
model.fit(x_train,y_train)
y_prediction = model.predict(x_test)

# Assuming y_prediction is your array of predictions
for i, pred in enumerate(y_prediction,1):
    print(f"{i}: {pred}")

#metrics--------------------------------------------------------------------------------

print('--------------------------------------------------------------------------')
print('r2_score:',r2_score(y_test,y_prediction))
print('mean_squared_error:',mean_squared_error(y_test,y_prediction))
print('mean_squared_error:',mean_absolute_error(y_test,y_prediction))
rmse = np.sqrt(mean_squared_error(y_test, y_prediction))
print("RMSE:", rmse)
print('--------------------------------------------------------------------------')

result = pd.DataFrame({
    'Actual_price':y_test,
    'prediction_price':y_prediction
})
print(result)

#visualization-------------------------------------------------------

result.plot(kind = 'bar',figsize=(10,5),color='skyblue')
plt.title('acutal_price vs prediction_price',fontsize = '14',color = 'green')
plt.xlabel('Numbers',fontsize = '14',color = 'red')
plt.ylabel('prediction,acutal',color = 'red')
plt.legend()
plt.tight_layout()
plt.show()

print('-------------------------------------------------------------------------')

#top 5 country
cal = 
```
